```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/admin/Documents/datasets/customer_churn.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
# drop customerID column because it is useless
df.drop('customerID',axis='columns',inplace=True)
```

```python
df.dtypes
```

```python
# dropping the rows which has totalCharges = " " Otherwise it will throw a error in pd.to_numeric()

df1 = df[df.TotalCharges != ' ']
```

```python
# converting totalCharges column data type from object to float
df1 = df1.astype({'TotalCharges' : 'float'})
```

```python
df1.dtypes
```

```python
df1.shape
```

```python
# Unique values for each column 
def print_unique_values_column(df):
    for column in df: 
        print(f'{df[column].dtypes} : {column} : {df[column].unique()}')
```

```python
print_unique_values_column(df1)
```

```python
# Let us replace all the value of 'No Internet Service' and 'No Phone Service' with No

df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_values_column(df1)
```

```python
# converting values of column which has Yes and No to 1 and 0 . 

yes_no_column = {'Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection',
                'TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn'}

for column in yes_no_column:
    df1[column].replace({'Yes' : 1 , 'No' : 0},inplace=True)

```

```python
df1.dtypes
```

```python
print_unique_values_column(df1)
```

```python
df1['gender'].replace({'Male' : 1 , 'Female' : 0},inplace=True)
```

```python
print_unique_values_column(df1)
```

```python
# one hot encoding for InternetService,Contract,PaymentMethod column
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
```

```python
print_unique_values_column(df2)
```

```python
# scaling data 

from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()

cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.head(2)
```

```python
# seperating input and output

x = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
# spliting test and train dataset

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
number_of_features = len(x_train.columns)
```

```python
number_of_features
```

```python
# deep learning model

import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(number_of_features,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile( optimizer = 'adam',
               loss = 'binary_crossentropy',
               metrics = ['accuracy'])

model.fit(x_train,y_train,epochs=1000)
```

```python
# evaluating test dataset
model.evaluate(x_test,y_test)
```

```python
# storing prediction result in an array
yp = model.predict(x_test)
yp[:5]
```

```python
# for every prediction value > 0.5 we will store 1 else store 0 in this new array
y_pred = []
for element in yp:
    if element > 0.5 :
        y_pred.append(1)
    else :
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
# importing libraries

from sklearn.metrics import confusion_matrix , classification_report
```

```python
# classification report 

print(classification_report(y_test,y_pred))
```

```python
# confusion matrix

import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
