```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/admin/Documents/datasets/sonar_dataset.csv',header = None)
df.sample(5)
```

```python
# checking whether there is any column which has null value

df.isna().sum()
```

```python
x = df.drop(60,axis="columns")
y = df[60]
```

```python
y
```

```python
# implementing one hot encoding for y and then deleting the first column

y = pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])

model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
# as you can see for train dataset it is showing accuracy of 100% 
# now let us see how it performs with test dataset

model.evaluate(x_test,y_test)
```

```python
# as you can see for test dataset it's accuracy is 0.7692 . This is a overfitting problem now let use solve
# this by adding a dropout layer after each hidden layer

model_dropout = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid'),
])

model_dropout.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])

model_dropout.fit(x_train,y_train,epochs=100,batch_size=8)

```

```python
# now for train data set accuracy is 0.88 let us check accuracy for test dataset

model_dropout.evaluate(x_test,y_test)
```

```python
# see now for test dataset accuracy is better than previous one with no dropout layer
```
