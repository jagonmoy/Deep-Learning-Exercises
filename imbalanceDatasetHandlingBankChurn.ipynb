```python
x = df.drop('Exited',axis='columns')
y = df['Exited']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state = 1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
y_pred = []
```

```python
def ann_model(x_train,y_train,x_test,y_test,epochs) :
    model = keras.Sequential([
        keras.layers.Dense(8,input_shape = (10,),activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
    ])
    model.compile( optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])

    model.fit(x_train,y_train,epochs=epochs)
    model.evaluate(x_test,y_test)
    yp = model.predict(x_test)
    for element in yp:
        if element > 0.5 :
            y_pred.append(1)
        else :
            y_pred.append(0)
```

```python
ann_model(x_train,y_train,x_test,y_test,1000)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

```python
# as you can see f1-score,recall,precision is way higher for 0 class than 1 class . 
# It's because of imbalance dataset . Now we will try to fix this issue with help of some techniques
```

## Under Sampling Technique

```python

df_class_0 = df[df['Exited'] == 0]
df_class_1 = df[df['Exited'] == 1]
```

```python
df_class_0.shape , df_class_1.shape
```

```python
# creating a new df class for zero class which is equals to one class 
df_class_0_under = df_class_0.sample(df_class_1.shape[0])
```

```python
df_class_0_under.shape
```

```python
# now we need to concat the df_class_0_under and df_class_1 for creating new dataframe which has equal amount of 
# zero class and one class value

df_under = pd.concat([df_class_0_under,df_class_1],axis=0)
```

```python
df_under.shape
```

```python
x = df_under.drop('Exited',axis='columns')
y = df_under['Exited']
```

```python
x.shape , y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 15, stratify = y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# as we can see equal number of 0 and 1 in our train dataset that is because you have used a parameter 'stratify' in
# our train_test_split method
y_train.value_counts()
```

```python
y_pred = []
```

```python
ann_model(x_train,y_train,x_test,y_test,1000)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

```python
# as you can see compare to previous result this time precision recall and f1 score for class 1 has improved 
# significantly
```

## Over Sampling Technique

```python
df_class_0.shape, df_class_1.shape
```

```python
# now we will create a new class for class 1 by increasing the range

df_class_1_over = df_class_1.sample(df_class_0.shape[0],replace=True)

df_class_1_over.shape
```

```python
df_over = pd.concat([df_class_0,df_class_1_over],axis=0)

df_over.shape
```

```python
x = df_over.drop('Exited',axis='columns')
y = df_over['Exited']
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=15,stratify = y)
```

```python
y_train.value_counts()
```

```python
y_pred = []
```

```python
ann_model(x_train,y_train,x_test,y_test,1000)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# better result than imbalanced dataset but similar to under sampling technique 
```

## SMOTE technique

```python
x = df.drop('Exited',axis='columns')
y = df['Exited']
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')

x_sm,y_sm = smote.fit_resample(x,y)
```

```python
y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sm,y_sm,test_size=0.2,random_state=15,stratify = y_sm)
```

```python
y_train.value_counts()
```

```python
y_pred = []
```

```python
ann_model(x_train,y_train,x_test,y_test,1000)
```

```python
print(classification_report(y_test,y_pred))
```

```python
## as you can see it is slighly better than under sampling and over sampling technique !
```
